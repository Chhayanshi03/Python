Q2: What is a scatter plot?Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
# A scatter plot is a type of graph used to display individual data points in a two-dimensional 
# coordinate system.

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

font1 = {'family':'serif', 'color': '#2E86C1','size':20}
color = 3 + np.random.normal(0, 2, 50)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('SCATTER PLOT',fontdict = font1)
plt.scatter(x,y, c = color)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
# The subplot() function in Matplotlib is used to create multiple plots within a single figure. 
# It allows you to divide the plotting area into a grid and specify which subplot a particular plot 
# should belong to.

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,4,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,4,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(1,4,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,4)

plt.suptitle('Use of subplot function')
plt.plot(x,y)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
#A bar plot is representation of data in form of bars. It can be horizontal as well as vertical.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

font1 = {'family':'serif', 'color': '#2E86C1','size':20}
font2 = {'family':'serif', 'color': '#6E2C00','size':15}
plt.xlabel('Company',fontdict = font2)
plt.ylabel('Profit',fontdict = font2)
plt.title('Company-Profit Representation',fontdict = font1)
plt.bar(company,profit, width = 0.4, color = '#AF7AC5')
plt.show()
```

```python
plt.ylabel('Company',fontdict = font2)
plt.xlabel('Profit',fontdict = font2)
plt.title('Company-Profit Representation',fontdict = font1)
plt.barh(company,profit, height = 0.4, color = '#82E0AA')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
# A Box plot is a way to visualize the distribution of the data by using a box and some vertical lines.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
plt.show()
```
